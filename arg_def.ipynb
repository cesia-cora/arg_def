```python
# Mejora de legibilidad en data1

# Grupo etario
data1.replace(to_replace='01.menor a 20 anios', value='Menor a 20 años', inplace=True)
data1.replace(to_replace='02.de 20 a 39 anios', value='De 20 a 39 años', inplace=True)
data1.replace(to_replace='03.de 40 a 49 anios', value='De 40 a 49 años', inplace=True)
data1.replace(to_replace='04.de 50 a 59 anios', value='De 50 a 59 años', inplace=True)
data1.replace(to_replace='05.de 60 a 69 anios', value='De 60 a 69 años', inplace=True)
data1.replace(to_replace='06.de 70 a 79 anios', value='De 70 a 79 años', inplace=True)
data1.replace(to_replace='07. de 80 anios y mas', value='De 80 años y más', inplace=True)
data1.replace(to_replace='08.sin especificar', value='Sin especificar', inplace=True)
```

### Visualización de datos

```python
plt.figure(figsize=(6, 3))
g = sns.catplot(data=data1, kind='count', x='region', hue='anio_def', height=4, aspect=2, palette='Spectral')
plt.title('Defunciones por año y región')
plt.xlabel('Región')
plt.ylabel('Cantidad')
g._legend.set_title('Año de defunción')
plt.xticks(
    rotation=25, 
    horizontalalignment='right',
    fontweight='light',
    fontsize=7  
)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
g = sns.catplot(data=data1, kind='count', x='anio_def', height=4, aspect=2, hue='grupo_etario', palette='Spectral')
plt.title('Defunciones por año')
plt.xlabel('Año de defunción')
plt.ylabel('Cantidad')
g._legend.set_title('Grupo etario')
plt.show()
```

### Cantidad de defunciones y porcentajes

```python
# Cantidad de defunciones en 2019, 2020 y 2021
cant_2019 = len(data1[data1['anio_def'] == 2019])
cant_2020 = len(data1[data1['anio_def'] == 2020])
cant_2021 = len(data1[data1['anio_def'] == 2021])

# Porcentajes
pri_porcentaje = ((cant_2020 - cant_2019)/cant_2019) * 100
seg_porcentaje = ((cant_2021 - cant_2020)/cant_2020) * 100
```

```python
defunciones = pd.DataFrame({'Año': [2019, 2020, 2021],'Cantidad de defunciones': [cant_2019, cant_2020, cant_2021]})
defunciones.style.set_properties(**{'background-color': 'coral', 'color': 'black', 'text-align': 'center'}).hide(axis='index')
```

```python
porcentajes = pd.DataFrame({'Período de tiempo': ['2019 a 2020', '2020 a 2021'], 'Cambio porcentual': [pri_porcentaje, seg_porcentaje]})
porcentajes.style.set_properties(**{'background-color': 'coral', 'color': 'black', 'text-align': 'center'}).hide(axis='index')
```

```python
# Defunciones en 2019, 2020, 2021 solo en la región "Centro"
data_2019 = data1[(data1['anio_def'] == 2019) & (data1['region'] == '1.Centro')]
data_2020 = data1[(data1['anio_def'] == 2020) & (data1['region'] == '1.Centro')]
data_2021 = data1[(data1['anio_def'] == 2021) & (data1['region'] == '1.Centro')]
```

```python
plt.figure(figsize=(30, 12))
g = sns.catplot(data=data_2020, kind='count', x='region', hue='grupo_causa_defuncion_CIE10', height=4, aspect=2)
plt.title('Defunciones en el año 2020')
plt.xlabel('Región: Buenos Aires, Córdoba, Santa Fe, Entre Ríos')
plt.ylabel('Cantidad')
g._legend.set_title('Grupo-causa por Clasificación Internacional de Enfermedades (CIE-10)')
plt.show()
```

```python
plt.figure(figsize=(30, 12))
g = sns.catplot(data=data_2021, kind='count', x='region', hue='grupo_causa_defuncion_CIE10', height=4, aspect=2)
plt.title('Defunciones en el año 2021')
plt.xlabel('Región: Buenos Aires, Córdoba, Santa Fe, Entre Ríos')
plt.ylabel('Cantidad')
g._legend.set_title('Grupo-causa por Clasificación Internacional de Enfermedades (CIE-10)')
plt.show()
```

### Cantidad de defunciones por enfermedad del sistema circulatorio

```python
def_cir_19 = len(data_2019[data_2019['grupo_causa_defuncion_CIE10'] == '0900 ENF  DEL SISTEMA CIRCULATORIO'])
def_cir_20 = len(data_2020[data_2020['grupo_causa_defuncion_CIE10'] == '0900 ENF  DEL SISTEMA CIRCULATORIO'])
def_cir_21 = len(data_2021[data_2021['grupo_causa_defuncion_CIE10'] == '0900 ENF  DEL SISTEMA CIRCULATORIO'])

cant_def_cir = pd.DataFrame({'Año': [2019, 2020, 2021], 'Cantidad de defunciones por enfermedad del sistema circulatorio': [def_cir_19, def_cir_20, def_cir_21]})
cant_def_cir.style.set_properties(**{'background-color': 'indianred', 'color': 'whitesmoke', 'text-align': 'center'}).hide(axis='index')
```

### Cantidad de defunciones por enfermedad del sistema respiratorio

```python
def_resp_19 = len(data_2019[data_2019['grupo_causa_defuncion_CIE10'] == '1000 ENF SISTEMA RESPIRATORIO'])
def_resp_20 = len(data_2020[data_2020['grupo_causa_defuncion_CIE10'] == '1000 ENF SISTEMA RESPIRATORIO'])
def_resp_21 = len(data_2021[data_2021['grupo_causa_defuncion_CIE10'] == '1000 ENF SISTEMA RESPIRATORIO'])

infecc_resp = pd.DataFrame({'Año': [2019, 2020, 2021], 'Cantidad de defunciones por enfermedad del sistema respiratorio': [def_resp_19, def_resp_20, def_resp_21]})
infecc_resp.style.set_properties(**{'background-color': 'skyblue', 'color': 'black', 'text-align': 'center'}).hide(axis='index')
```

### Cantidad de casos de infecciones respiratorias

```python
# Datasets separadas por año
año_2018 = data2[data2['anio'] == 2018]
año_2019 = data2[data2['anio'] == 2019]
año_2020 = data3[data3['año'] == 2020]
año_2021 = data3[data3['año'] == 2021]
año_2022 = data4[data4['año'] == 2022]

# Suma de cantidades de infecciones
val_cant_2018 = año_2018['cantidad_casos'].sum()
val_cant_2019 = año_2019['cantidad_casos'].sum()
val_cant_2020 = año_2020['cantidad_casos'].sum()
val_cant_2021 = año_2021['cantidad_casos'].sum()
val_cant_2022 = año_2022['cantidad_casos'].sum()

cant_infecc_resp = pd.DataFrame({'Año': [2018, 2019, 2020, 2021, 2022], 'Cantidad de casos de infección respiratoria': [val_cant_2018, val_cant_2019, val_cant_2020, val_cant_2021, val_cant_2022]})
cant_infecc_resp.style.set_properties(**{'background-color': 'teal', 'color': 'whitesmoke', 'text-align': 'center'}).hide(axis='index')
```
